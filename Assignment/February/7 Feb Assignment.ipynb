```python
# 1.  create a python function that will check whether the password entered by the user is 
#     correct or not.  
#     1. The Password should contain at least two uppercase letters and at least two lowercase letters.
#     2. The Password should contain at least a number and three special characters.
#     3. The length of the password should be 10 characters long.

# Program Starts
password=input("Enter Password: ")

lower_case=list(filter(lambda x: x>='a' and x<='z', password))
upper_case=list(filter(lambda x: x>='A' and x<='Z', password))
number=list(filter(lambda x: x>='0' and x<='9', password))
special_case=list(filter(lambda x: ord(x)>=33 and ord(x)<=47 or ord(x)>=58 and ord(x)<=64 , password))

if(len(password)>=10 and len(lower_case)>=2 and len(upper_case)>=2 and len(number)>=1 and len(special_case)>=3):
   print("Valid Password")
else:
   print("Invalid Password")
```
